# Taylor's Theorem

**Definition**

如果函数 $f(x)$ 在点 $x=a$ 处具有任意阶导数，则它在该点的泰勒级数定义为：$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x-a)^n$$$f^{(n)}(a)$：表示函数在 $a$ 点的 $n$ 阶导数。

$n!$：阶乘，作为归一化系数


> 当对 $x^n$ 求导时：
  
> 一次导数：$n \cdot x^{n-1}$

> 二次导数：$n(n-1) \cdot x^{n-2}$

> $n$ 次导数：$n!$

> 为了让多项式的第 $n$ 阶导数等于函数的第 $n$ 阶导数，需要除以 $n!$ 来把求导产生的系数归一化



**麦克劳林级数** (Maclaurin Series)是泰勒级数的特例。当我们选择在原点 $a=0$ 展开时，公式简化为：$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!} x^n$$



---




**example1:    $e^x = \sum \frac{x^n}{n!}$**


$f^{(n)}(x) = e^x$, $f^{(n)}(0) = 1$ (所有阶导数在 0 点的值都是 1)

将 $f^{(n)}(0) = 1$ 代入麦克劳林公式：$$e^x = \sum_{n=0}^{\infty} \frac{1}{n!} x^n = 1 + \frac{x}{1!} + \frac{x^2}{2!} + \frac{x^3}{3!} + \dots$$

$$\sum_{n=0}^{\infty} \frac{x^n}{n!} = e^x$$



---



**example2: 拆分：$e^{\lambda} = 1 + \sum_{j=1}^{\infty} \dots$**

$$e^{\lambda} = \sum_{j=0}^{\infty} \frac{\lambda^j}{j!} = \underbrace{\frac{\lambda^0}{0!}}_{\text{当 } j=0} + \sum_{j=1}^{\infty} \frac{\lambda^j}{j!}$$

$$e^{\lambda} = 1 + \sum_{j=1}^{\infty} \frac{\lambda^j}{j!}$$
