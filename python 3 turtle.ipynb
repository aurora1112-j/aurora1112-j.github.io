## 绘制开花的树

```python
import turtle

toplevel = 8  # 一共递归6层
angle = 30
rangle = 15

def drawTree(length, level):
    turtle.left(angle)  # 绘制左枝
    turtle.color("black")
    turtle.forward(length)

    if level == toplevel:  # 叶子
        turtle.color("pink")
        turtle.circle(2, 360)

    if level < toplevel:  # 在左枝退回去之前递归
        drawTree(length - 10, level + 1)
    turtle.back(length)

    turtle.right(angle + rangle)  # 绘制右枝
    turtle.color("black")
    turtle.forward(length)

    if level == toplevel:  # 叶子
        turtle.color("pink")
        turtle.circle(2, 360)

    if level < toplevel:  # 在右枝退回去之前递归
        drawTree(length - 10, level + 1)
        turtle.color("black")
    turtle.back(length)
    turtle.left(rangle)

turtle.left(90)
turtle.penup()
turtle.back(300)
turtle.pendown()
turtle.forward(100)
turtle.speed('fastest')
drawTree(80, 1)

turtle.done()
```

## 绘制中国国旗

```python
# 我们将使用turtle模块来绘制中国的五星国旗。国旗的规格有严格的比例规定：

# - 国旗的长宽比为3:2

# - 五颗星星的位置和大小也有固定比例

# 根据《中华人民共和国国旗法》：

# 国旗的旗面为红色，长方形，其长与高为三与二之比，旗面左上方缀黄色五角星五颗。

# 一星较大，其外接圆直径为旗高十分之三，居左；

# 四星较小，其外接圆直径为旗高十分之一，环拱于大星之右。

# 四颗小星都有一角正对大星的中心点。

# 我们按照以下步骤绘制：

# 1. 设置画布大小为宽900像素，高600像素（3:2）。

# 2. 绘制红色矩形作为旗面。

# 3. 绘制大五角星：位于旗面左上四分之一区域的正中（具体位置有规定）。

# 4. 绘制四颗小五角星：每颗小星的中心点都在以大星中心为原点的某条半径上，且小星的一个角指向大星中心。

# 注意：实际绘制时，我们按照比例计算位置和大小。

# 具体位置（以旗面左上角为原点，向右为x正方向，向下为y正方向）：

# 大五角星：从旗面左上角，向下移动旗高的1/10，向右移动旗长的1/4处（但实际规定大星中心点在矩形左上角区域，我们按照标准：大星中心点位于（旗长的1/4，旗高的1/4）处？）

# 实际上，根据标准：大星的中心点在旗面的左上方1/4区域的中心，即（旗长*1/4，旗高*1/4）？但是查阅资料，大星的中心点位于（旗长的1/4，旗高的1/4）处。

# 然而，根据《国旗法》说明：大五角星的中心点在旗面左上角（以旗面左上角为原点）的（5/16, 5/16）处（这个比例是我根据常见画法总结的，因为旗高3等分，旗长5等分，但实际常见画法如下）：

# 另一种常见的比例：大星中心点位于（旗长1/4，旗高1/4）处，但这样会太靠左上，所以通常调整。

# 根据中华人民共和国国家标准GB 12983-2004《国旗尺寸和颜色》：

# 大五角星的中心点：在长方形上五下五、左五右十之处（将旗面长分为20等份，高分为20等份）。具体位置：

# 大星中心点：左上角（0,0）为起点，则（5,5）处（即长和高的1/4处，因为5/20=1/4）？但是这样计算：旗长20等份，每份为旗长的1/20，旗高20等份，每份为旗高的1/20。

# 大星中心点：位于（5,5）处（即从左上角开始，向右5份，向下5份）。

# 四颗小星的中心点：

# 第一颗小星：位于（10,2）处（向右10份，向下2份）

# 第二颗小星：位于（12,4）处（向右12份，向下4份）

# 第三颗小星：位于（12,7）处（向右12份，向下7份）

# 第四颗小星：位于（10,9）处（向右10份，向下9份）

# 五角星的外接圆直径：

# 大星：直径占旗高的3/10（即6份，因为旗高20份，6份就是旗高的3/10）

# 小星：直径占旗高的1/10（即2份）

# 注意：以上坐标和尺寸都是以旗面划分为20*20的网格为基础的。

# 因此，我们设定：

# 旗长（width）= 900

# 旗高（height）= 600

# 则每份的宽度：dx = 900/30？不对，标准是长20份，高20份。但我们的旗长是900，旗高600，所以每份在长度方向为900/20=45，在高度方向为600/20=30。

# 但是，这样会导致网格不是正方形的，所以我们需要调整吗？实际上，我们绘制五角星时，外接圆直径按旗高的比例来，位置按网格来。

# 步骤：

# 1. 绘制红色背景矩形。

# 2. 计算网格：将旗长分为20等份，每份45；旗高分为20等份，每份30。

# 3. 绘制大五角星：中心点(5*45, 5*30)，外接圆半径 = (旗高 * 3/10) / 2 = (600 * 0.3) / 2 = 90

# 4. 绘制四颗小五角星：中心点分别为(10*45, 2*30), (12*45,4*30), (12*45,7*30), (10*45,9*30)，外接圆半径 = (600 * 1/10) / 2 = 30

# 5. 五角星的绘制：我们写一个函数，给定中心坐标、外接圆半径、旋转角度（小星需要旋转，使其一个角指向大星中心）。

# 注意：小星的旋转角度：每个小星的中心点与大星中心点（5*45,5*30）的连线与水平线的夹角，这个夹角就是小星需要旋转的角度（但注意五角星的一个角要指向大星中心，所以旋转角度就是这个夹角）。

# 计算夹角（以弧度或角度）：

# 设小星中心为(x1,y1)，大星中心为(x0,y0)，则向量为(x0-x1, y0-y1)？注意坐标系的y轴向下是正方向，所以角度需要调整。

# 实际上，我们使用math.atan2(dy, dx)来计算角度，其中dy = y0-y1, dx = x0-x1，然后转换为角度。

# 但是，注意turtle绘图的0度方向是向右，逆时针为正。而我们希望小星的一个角指向大星中心，因此需要将五角星旋转这个角度。

# 然而，五角星本身的绘制是从顶点开始，所以我们需要调整起始角度。

# 五角星每个顶点的角度间隔为72度。通常，我们绘制五角星时，从最上面的顶点开始（即12点钟方向），但为了旋转，我们可以先绘制一个顶点向上的五角星，然后根据所需方向旋转。

# 所以，对于每颗小星，我们计算从该小星中心指向大星中心的方向向量，然后计算这个向量的角度（从x轴正方向逆时针旋转的角度），然后让五角星的一个角指向这个方向。由于五角星的一个顶点原本是向上的（即90度方向），所以我们需要将五角星旋转（计算得到的角度-90度）？这样，原本向上的顶点就会指向大星方向。

# 但注意：我们计算的是从小星中心指向大星中心的方向，这个方向的角度为θ，那么我们要让五角星的一个顶点指向这个方向，而原本顶点是90度方向（即垂直向上），所以旋转角度为θ-90度（如果θ是相对于x轴正方向的角度）。

# 步骤总结：

# 1. 设置画布和turtle参数。

# 2. 画红色矩形。

# 3. 画大五角星（不旋转，起始角度为90度，即顶点向上）。

# 4. 对每颗小星：

# a. 计算小星中心到大星中心的向量（dx, dy）= (大星x-小星x, 大星y-小星y)

# b. 计算向量角度（单位：度）θ = math.degrees(math.atan2(dy, dx))

# c. 绘制五角星时，旋转角度设为θ-90（这样，原本向上的顶点就会指向大星方向）

# 注意：turtle的setheading(angle)设置方向，但我们在绘制五角星函数内部控制旋转。

# 由于turtle绘图是顺序执行，我们需要控制抬笔、落笔和移动。

# 我们将编写一个绘制五角星的函数：

# def draw_star(x, y, radius, rotation=0, color="yellow"):

# turtle.penup()

# turtle.goto(x, y)

# turtle.setheading(rotation)  # 设置起始方向，指向旋转后的顶点

# turtle.pendown()

# turtle.fillcolor(color)

# turtle.begin_fill()

# for i in range(5):

# turtle.forward(radius * math.sin(math.pi/5*2))  # 五角星边长与外接圆半径的关系：边长 = 2 * radius * sin(36°)

# turtle.right(144)   # 内角144度

# turtle.end_fill()

# 但是，上面的方法绘制五角星时，顶点方向由rotation控制，我们设定rotation为θ-90，这样顶点就会指向大星。

# 然而，上面的边长公式：五角星的每个顶点在外接圆上，相邻顶点间的圆心角为72度。从一个顶点到下一个顶点的弦长，根据余弦定理：边长 = 2 * radius * sin(36°)，因为圆心角72度的一半是36度。

# 但是，我们也可以采用另一种方法：每次旋转144度（因为五角星的内角是144度）。

# 注意：我们绘制的是五角星轮廓（五边形），不是实心五边形？但用turtle填充即可。

# 由于turtle的坐标系统：原点在画布中心，但我们的旗面原点在左上角，所以我们需要调整坐标计算。但为了方便，我们可以将turtle的坐标系设置为以左上角为原点，x向右，y向下。或者，我们通过计算将每个点的坐标转换为turtle坐标（画布中心为原点）。通常，我们设置画布大小为900x600，那么左上角的坐标是(-450, -300)（如果以中心为原点）。

# 因此，我们绘制矩形时，从左上角(-450, -300)开始。

# 步骤：

# 1. 导入turtle和math

# 2. 设置窗口大小

# 3. 计算各个点的坐标（转换为turtle坐标系）

# 转换方法：

# turtle坐标系：中心为(0,0)，左上角坐标为(-width/2, -height/2)

# 我们之前计算的网格坐标（x_grid, y_grid）在旗面坐标系（左上角为原点，x向右，y向下）中，那么对应的turtle坐标：

# x = -width/2 + x_grid * (width/20)   # 注意：网格每份在x方向占45，y方向占30

# y = -height/2 + y_grid * (height/20)   # 但是注意：旗面坐标系中y向下为正，而turtle坐标系中y向上为正，所以我们要取负？

# 不对，我们直接按旗面坐标系绘制：在turtle中，旗面的左上角坐标为(-450, -300)，那么旗面坐标系的点(x_grid, y_grid)在turtle坐标系中为：

# x = -450 + x_grid

# y = -300 + y_grid

# 但是，我们的网格单位是：x_grid是45的倍数，y_grid是30的倍数。

# 因此，对于大星中心点(5,5)在旗面坐标系中的位置为（5*45, 5*30）=(225,150)，在turtle坐标系中：

# x = -450 + 225 = -225

# y = -300 + 150 = -150

# 但是，注意：turtle的y轴向上为正，而我们旗面坐标系中y向下为正，所以上面计算出的y值实际上是向下的，在turtle中就是向上移动了？不对，我们旗面坐标系中，从左上角(0,0)点，向下移动150，在turtle坐标系中，初始左上角(-450,-300)，向下移动150，那么y坐标从-300变为-150（因为turtle中向上移动150，从-300到-150是向上移动）。这不符合我们的预期。

# 因此，我们需要将旗面坐标系的y坐标转换为turtle坐标系的y坐标：旗面坐标系的点(x,y)（其中y是向下为正）在turtle坐标系中y值应该是：-(y) + (height/2) ？不对，这样会很乱。

# 另一种方法：我们绘制整个旗面时，将旗面的左上角设为(-450, 300)（即y轴向上为正，那么旗面从y=300到y=-300，这样向下移动就是y值减小）。这样，旗面坐标系中(x, y)（y向下）在turtle坐标系中就是：

# x_turtle = -450 + x

# y_turtle = 300 - y   # 因为旗面坐标系y向下，所以当y=0时，turtle的y=300；当y=600时，turtle的y=-300

# 这样，我们定义：

# x_turtle = -450 + x_grid * 45

# y_turtle = 300 - y_grid * 30

# 例如，大星中心点(5,5)在旗面坐标系中x=225, y=150，那么turtle坐标：x_turtle=-450+225=-225, y_turtle=300-150=150

# 因此，我们按此转换所有点。

# 但是，这样会使得整个旗面倒过来吗？不会，因为我们绘制矩形时也要按这个坐标系。

# 绘制矩形：

# turtle.penup()

# turtle.goto(-450, 300)   # 左上角

# turtle.pendown()

# turtle.color("red")

# turtle.begin_fill()

# turtle.goto(450, 300)    # 右上角

# turtle.goto(450, -300)   # 右下角（注意：在turtle中，y=-300）

# turtle.goto(-450, -300)

# turtle.goto(-450, 300)

# turtle.end_fill()

# 然后，我们按照转换公式计算五角星的中心点turtle坐标。
```

```python
import turtle
import math

# 设置国旗尺寸参数
WIDTH = 900       # 国旗宽度
HEIGHT = WIDTH * 2 // 3  # 国旗高度（3:2比例）

# 设置颜色
RED = "#de2910"   # 中国红
YELLOW = "#ffde00" # 五星黄

# 创建画布和画笔
screen = turtle.Screen()
screen.title("中华人民共和国国旗")
screen.setup(WIDTH + 100, HEIGHT + 100)
screen.bgcolor("white")

pen = turtle.Turtle()
pen.speed(10)  # 设置绘制速度为最快
pen.hideturtle()

# 绘制红色背景
pen.penup()
pen.goto(-WIDTH/2, HEIGHT/2)
pen.pendown()
pen.color(RED)
pen.begin_fill()
for _ in range(2):
    pen.forward(WIDTH)
    pen.right(90)
    pen.forward(HEIGHT)
    pen.right(90)
pen.end_fill()

# 绘制大五角星（位置在左上四分之一区域）
def draw_star(x, y, radius, rotation, color=YELLOW):
    """绘制五角星
    
    参数:
    x, y: 五角星中心坐标
    radius: 外接圆半径
    rotation: 旋转角度（度）
    color: 填充颜色
    """
    pen.penup()
    pen.goto(x, y)
    pen.setheading(rotation)
    pen.forward(radius)
    pen.setheading(rotation + 180 - 18)  # 调整起始位置
    
    pen.pendown()
    pen.color(color)
    pen.begin_fill()
    
    # 绘制五角星
    for _ in range(5):
        pen.forward(2 * radius * math.sin(math.pi/5))
        pen.right(144)  # 五角星内角144度
        
    pen.end_fill()

# 计算五角星位置和大小
# 大五角星
big_star_radius = HEIGHT * 3 / 20  # 大星外接圆直径 = 旗高3/10
big_star_x = -WIDTH/2 + WIDTH * 5/20  # 左起1/4处
big_star_y = HEIGHT/2 - HEIGHT * 5/20  # 上起1/4处

# 绘制大五角星（指向正上方）
draw_star(big_star_x, big_star_y, big_star_radius, 90)

# 四颗小五角星的位置（相对于大星中心）
small_star_radius = HEIGHT / 20  # 小星外接圆直径 = 旗高1/10
star_positions = [
    # (相对大星的x偏移, 相对大星的y偏移, 旋转角度)
    (3, -1, math.degrees(math.atan2(1, 3))),   # 第一颗小星
    (5, -2, math.degrees(math.atan2(2, 5))),   # 第二颗小星
    (5, -5, math.degrees(math.atan2(5, 5))),   # 第三颗小星
    (3, -6, math.degrees(math.atan2(6, 3)))    # 第四颗小星
]

# 计算单位长度（将旗面分为20等分）
unit_x = WIDTH / 20
unit_y = HEIGHT / 20

# 绘制四颗小五角星
for dx, dy, angle in star_positions:
    # 计算小星中心位置
    x = big_star_x + dx * unit_x
    y = big_star_y - dy * unit_y
    
    # 计算小星指向大星中心的旋转角度
    rotation = 90 - angle  # 调整为turtle坐标系
    
    # 绘制小五角星
    draw_star(x, y, small_star_radius, rotation)

# 添加文字说明
pen.penup()
pen.goto(0, -HEIGHT/2 - 40)
pen.color("black")
pen.write("中华人民共和国国旗", align="center", font=("SimHei", 24, "bold"))

# 完成绘制
turtle.done()

# 输出结果，小星星的位置不对
```
