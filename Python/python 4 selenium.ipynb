## 要循环抓取怎么办？

```python
#注意缩进关系、英文要准确
movies = []
i= 0
while i<10:
    movies_list = driver.find_elements(By.TAG_NAME,"Li")
    # "Li" 是一个定位策略，用于查找HTML文档中所有具有特定标签名的元素。在这个例子中，它用于查找所有<li>标签的元素。
    for element in movies_list:
        try:
            title = element.find_element(By.CLASS_NAME, "title").text
            rating = element.find_element(By.CLASS_NAME, "rating_num").text
            movie = {"title":title,"rating":rating}
            movies.append(movie)
        except:
            pass
    driver.find_element(By.CLASS_NAME, "next").click()
    driver.switch_to.window(driver.window_handles[0])
    i=i+1
else:
    print(movies)
    
```

```python
# 另一种写法。https://movie.douban.com/top250 这个网页的特点
# 翻到第二页，https://movie.douban.com/top250?start=25&filter=
# 翻到第三页，https://movie.douban.com/top250?start=50&filter=
# 规律：25 的倍数
# 通过引入str()函数来登录 网址 
```

```python
#注意缩进关系、英文要准确。
movies = []
i= 0
while i<10:
    driver.get("https://movie.douban.com/top250?start="+str(i*25))  # 注意双引号的位置
    movies_list = driver.find_elements(By.TAG_NAME,"Li")        
    for element in movies_list:
        try:
            title = element.find_element(By.CLASS_NAME, "title").text
            rating = element.find_element(By.CLASS_NAME, "rating_num").text
            movie = {"title":title,"rating":rating}
            movies.append(movie)
        except:
            pass
    i=i+1
else:
    print(movies)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
```

```python
def get_douban_group_posts(driver, url):
    driver.get(url)
    posts = []

    try:
        # 等待页面加载完成
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "olt"))
        )

        # 获取帖子列表
        table = driver.find_element(By.CLASS_NAME, 'olt')
        for row in table.find_elements(By.TAG_NAME, 'tr')[1:]:  # 跳过标题行
            cols = row.find_elements(By.TAG_NAME, 'td')
            if cols:
                post = {
                    '标题': cols[0].text.strip(),
                    '作者': cols[1].text.strip(),
                    '回应': int(cols[2].text.strip()),
                    '时间': cols[3].text.strip()
                }
                posts.append(post)

    except TimeoutException:
        print("Loading took too much time!")

    return posts

def scrape_douban_group(url, pages=4):
    # 设置Selenium驱动程序，这里使用Chrome
    driver = webdriver.Chrome()
    all_posts = []
    for page in range(pages):
        print(f"Fetching page {page}")
        posts = get_douban_group_posts(driver, url)
        all_posts.extend(posts)
        
        # 找到并点击下一页按钮
        try:
            next_button = driver.find_element_by_link_text('后页')
            if next_button:
                next_button.click()
        except:
            print("No next page or next page button not found.")
            break

    driver.quit()
    return all_posts

if __name__ == '__main__':
    url = 'https://www.douban.com/group/655381/discussion?start=0&type=elite'
    posts = scrape_douban_group(url)
    for post in posts:
        print(post)
```
