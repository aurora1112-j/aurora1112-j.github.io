```python
from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker
 
 
# 统计地图构建数据时， 地区名要跟地图上的名字完全对应，不可以增减字数。
# 统计地图的地图类型选取，需要注意：
# 1. 中国地图传入: 'china'
# 2. 外国地图传入: '美国', '英国', '俄罗斯'
# 3. 中国省份和城市: '广东', '梅州', '新疆'
 
 
#各省观看易烊千玺电影《送你一朵小红花》的人数
 
data =  [
         ['江西省', 68000],
         ['福建省', 80000], ['广东省', 82000], ['湖南省', 65000], ['广西省', 65000],
         ['贵州省', 30000], ['河南省', 50000], ['湖北省', 70000], ['河北省', 68000],
         ['陕西省', 7000], 
        ]
 
#创建地图对象 
map_chart = Map()
map_chart.set_global_opts(
    title_opts=opts.TitleOpts(title="《送你一朵小红花》观影人数"),
    visualmap_opts=opts.VisualMapOpts(max_=500)
)
map_chart.add("观影人数", data, "china"
             ,zoom=1.5 # 设置缩放
             ,center=[100, 36] # 设置默认中心位置
             ,is_roam=False    # 禁用缩放 拖拽
             )
map_chart.render_notebook()




# 也可以写在以下，地图可以缩放和自由拖拽
# c = (
#     Map()
#     .add("观影人数", data, "china") 
#     .set_global_opts(
#         title_opts=opts.TitleOpts(title="Map-中国地图"), visualmap_opts=opts.VisualMapOpts()
#     )
# )
# c.render_notebook()




```

```python
from pyecharts.charts import Map
from pyecharts import options as opts

# 创建一个 Map 对象
malaysia_map = Map(init_opts=opts.InitOpts(width="800px", height="600px"))

# 添加马来西亚地图
malaysia_map.add(
    series_name="马来西亚",
    data_pair=[("马来西亚", 1)],
    maptype="马来西亚",
    label_opts=opts.LabelOpts(is_show=True),
)

# 设置全局选项
malaysia_map.set_global_opts(
    title_opts=opts.TitleOpts(title="马来西亚地图"),
    visualmap_opts=opts.VisualMapOpts(is_show=False),
)

# 渲染地图
malaysia_map.render("malaysia_map.html")
```

```python
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.globals import ChartType

geo = Geo(init_opts=opts.InitOpts(width="800px", height="600px"))

# 添加坐标点（Ashgabat 是土库曼斯坦首都）
geo.add_schema(maptype="world")
geo.add_coordinate("Ashgabat", 58.3833, 37.9500)  # 经度, 纬度

geo.add(
    series_name="Turkmenistan",
    data_pair=[("Ashgabat", 100)],
    type_=ChartType.EFFECT_SCATTER,
    label_opts=opts.LabelOpts(is_show=True),
)

geo.set_global_opts(title_opts=opts.TitleOpts(title="土库曼斯坦地图（世界模式）"))
geo.render("turkmenistan_geo.html")
```

```python
# 越南地图，显示胡志明和河南两座城市 
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType

# 1. 手动添加越南城市的经纬度（经度, 纬度）
coords = {
    "Hanoi": (105.8342, 21.0278),          # 河内
    "Ho Chi Minh City": (106.6297, 10.8231),  # 胡志明市
}

# 2. 数据准备（城市 + 颜色数值）
data = [
    ("Hanoi", 100),          # 红色
    ("Ho Chi Minh City", 200),  # 黄色
]

# 3. 创建 Geo 图表
geo = Geo(init_opts=opts.InitOpts(width="800px", height="600px"))

# 4. 设置越南地图（如果 "越南" 不生效，可以尝试 "Vietnam"）
geo.add_schema(maptype="越南")

# 5. 手动添加坐标（避免 PyEcharts 找不到位置）
for city, coord in coords.items():
    geo.add_coordinate(city, coord[0], coord[1])

# 6. 添加数据点（只显示 Hanoi, Ho Chi Minh City）
geo.add(
    series_name="越南重点城市",
    data_pair=data,
    type_=ChartType.EFFECT_SCATTER,  # 特效散点图
    symbol_size=12,
    label_opts=opts.LabelOpts(is_show=True, formatter="{b}"),  # 显示城市名称
)

# 7. 设置视觉映射（控制颜色）
geo.set_global_opts(
    title_opts=opts.TitleOpts(title="越南重点城市标注"),
    visualmap_opts=opts.VisualMapOpts(
        is_show=True,
        min_=100,
        max_=300,
        range_color=["red", "yellow"],  # 颜色映射
        is_piecewise=True,  # 分段显示
        pieces=[
            {"min": 100, "max": 100, "label": "Hanoi", "color": "red"},
            {"min": 200, "max": 200, "label": "Ho Chi Minh City", "color": "yellow"},
        ],
    ),
)

# 8. 渲染成 HTML 文件
geo.render("vietnam_cities.html")
```

## 根据geoBoundaries 数据绘制地图

```python
# pip install geopandas matplotlib shapely
# 下载geoBoundaries数据 https://www.geoboundaries.org/countryDownloads.html
#
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Windows显示中文的语法
plt.rcParams['font.sans-serif']=['SimHei']

# 读取 GeoJSON 或 Shapefile 数据
boundary_file_path = 'geoBoundaries-TKM-ADM1_simplified.geojson'  # 替换为你的文件路径
gdf = gpd.read_file(boundary_file_path)

# 可选：选择特定的列或过滤数据
# gdf = gdf[['ADM0_NAME', 'ADM1_NAME', 'geometry']]  # 替换为你需要的列名

# 绘制地图
fig, ax = plt.subplots(figsize=(12, 8))
gdf.boundary.plot(ax=ax, edgecolor='black', linewidth=0.5)  # 绘制边界
gdf.plot(ax=ax, color='lightblue', alpha=0.5, edgecolor='black')  # 绘制面

# 添加标题和标签
plt.title('GeoBoundaries Map')
plt.xlabel('经度')
plt.ylabel('纬度')

# 添加国家或行政区划名称
for idx, row in gdf.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, row['ADM0_NAME'], 
             ha='center', va='center', fontsize=8)

# 显示地图
plt.show()
```
