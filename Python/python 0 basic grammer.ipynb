## 字典 

```python
#字典是映射类型的数据，使用花括号创建，包括一系列键，每个键对应一个值，键和值之间用冒号连接。键是唯一的，值不唯一。
#字典是可变的、无序的数据组合，可以添加或移除数据项。
```

```python
# 创建一个空字典  
empty_dict = {}  
```

```python
# 创建一个包含键值对的字典 
mydict = {1:'Emma',2:'Mike',4:'Lily',5:'Robert'}
print (mydict.keys())  #输出字典变量dict的全部键
```

```python
# 创建一个包含键值对的字典 
mydict = {1:'Emma',2:'Mike',4:'Lily',5:'Robert'}
value = mydict[1] # 访问键为1的值
print (value) 
```

```python
# 案例
my_dict = {  
    "name": "Alice",  
    "age": 30,  
    "city": "New York"  
}
value = my_dict["name"] 
print(value) 
```

```python
#通过键来修改字典中的值。
my_dict["age"] = 31
print(my_dict)
```

```python
# 添加字典，新的键值对  
my_dict["email"] = "alice@example.com"  
print(my_dict) 
```

```python
# 删除"email"的键值对
del my_dict["email"]    
print(my_dict)
```

```python
#如果访问的键不存在于字典中，Python会抛出KeyError异常。
#为避免这种情况，可以使用get()方法，它会在键不存在时返回None或指定的默认值。
```

```python
value = my_dict.get("address")  # 键"address"不存在，返回None  
print(value)  # 输出: None  
  
value = my_dict.get("address", "No address provided")  # 返回指定的默认值  
print(value)  # 输出: No address provided
```

```python
# 遍历字典的键  
for key in my_dict:  
    print(key)  # 输出: name, city  
```

```python
for value in my_dict.values():  
    print(value) 
```

```python
for key, value in my_dict.items():  
    print(key, value) 
```

## 循环命令

```python
#允许多次执行一条或多条语句。常见的循环命令有for语句和while语句。符合条件执行，不符合返回其他结果。
x = [2,4,6,8,10]   # 创建列表X，列表为序列数据
for number in (x):
    print (number)
```

### for语句

```python
# for常与else连用，表示在结束循环时执行else语句。注意格式有缩进，注意英文冒号
x =  [2, 4, 6, 8, 10] 
for number in (x):
    if number < 5:
        print (number,'小于5')
    else:
        print (number,'大于或等于5')
```

### While语句

```python
#while语句用于执行循环程序，处理需要重复执行的任务，例如抓取网页数据时需要翻页，就需要用到循环语句，
#注意格式的缩进和英文冒号
X = 0
while X < 9:
    print (X, '小于9') #判断条件为X小于9
    X = X + 1   #符合条件则执行输出语句，循环命令
else:
    print(X,'大于或等于9')
```
