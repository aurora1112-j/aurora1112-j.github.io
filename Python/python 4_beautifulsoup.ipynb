```python
import requests
from bs4 import BeautifulSoup
# Make a GET request to the target URL
url = 'https://media.nju.edu.cn/campus_prize/dyj/list.htm'
response = requests.get(url)
# Check the response status code to make sure the request was successful

if response.status_code == 200:
    # Parse the HTML content of the response using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the links on the page and print their URLs
    links = soup.find_all('a')
    for link in links:
        print(link.get('href'))
else:
    # If the request was not successful, print an error message
    print('Error: Could not retrieve data')
```

# 抓取网页数据

```python
import requests
from bs4 import BeautifulSoup

# URL of the HTML page containing the dataset
url = "https://www.stats.gov.cn/sj/pcsj/rkpc/5rp/html/l0605a.htm"

# Send a GET request to fetch the HTML content of the page
response = requests.get(url)
response.encoding = 'GBK'  # 指定网页编码为GBK  

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table containing the dataset
    table = soup.find('table')
    
    # Check if the table was found
    if table:
        # Extract data from the table
        rows = table.find_all('tr')
        
        # Loop through each row
        for row in rows:
            # Extract data from each cell in the row
            cells = row.find_all('td')
            
            # Print data from each cell
            for cell in cells:
                print(cell.text.strip(), end='\t')
            print()  # Move to the next row
    else:
        print("No table found on the page.")
else:
    print("Failed to fetch the page. Status code:", response.status_code)
```

# 抓取豆瓣数据

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_douban_group_posts(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    posts = []
    table = soup.find('table', class_='olt')
    if table:
        for row in table.find_all('tr')[1:]:  # 跳过标题行
            cols = row.find_all('td')
            if cols:
                post = {
                    '标题': cols[0].find('a').get_text(strip=True) if cols[0].find('a') else '',
                    '回应': int(cols[2].get_text(strip=True)) if cols[2] else 0,
                    '最后回应时间': cols[3].get_text(strip=True) if cols[3] else ''
                }
                posts.append(post)
    return posts

def scrape_douban_group(url, pages=4):
    all_posts = []
    for page in range(pages):
        print(f"Fetching page {page}")
        posts = get_douban_group_posts(url, start=page * 25)  # 假设每页显示25条帖子
        all_posts.extend(posts)
    return all_posts

if __name__ == '__main__':
    url ='https://www.douban.com/group/655381/discussion?start=0&type=elite'
    posts = get_douban_group_posts(url)
    for post in posts:
        print(post)
```
