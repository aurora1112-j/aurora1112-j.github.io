### 案例2：阅读行为对主观幸福感/心理健康的影响（D30）

```python
df["A30_4"]
```

```python
# 阅读行为
# A30.过去一年，您是否经常在空闲时间从事以下活动？【出示示卡3】
# 每天 一周数次 一月数次 一年数次或更少 从不 不知道 拒绝回答
# 4. 读书 /报纸 /杂志 1 2 3 4 5 98 99
df["A30_4_sorted"] = df["A30_4"].cat.reorder_categories(['不知道', '从不', '一年数次或更少', '一月数次', '一周数次', '每天'])
df["A30_4_sorted"]
```

```python
df["A30_4_float"] = df["A30_4"]\
.replace("从不", 1)\
.replace("一年数次或更少", 2)\
.replace("一月数次", 3)\
.replace("一周数次", 4)\
.replace("每天", 5)\
.replace("不知道", 0)
df["A30_4_float"] = df["A30_4_float"].astype(float)
df["A30_4_float"]
```

```python
df["D36"]
```

```python
# 主观幸福感
# D36.请给您目前的幸福感评分（最高 10分，最低 0分）
# 记录：[____|____]分
# 【访员注意】98.不知道、99.拒绝回答
df["D36_float"] = df["D36"]\
.replace("不知道", 0)\
.replace("拒绝回答", 0).astype(float)
df["D36_float"]
```

```python
# D24.总的来说，您认为您的健康状况如何（这里说的健康包括生理和心理健康）？
df["D24_float"] = df["D24"]\
.replace("差", 1)\
.replace("一般", 2)\
.replace("好", 3)\
.replace("很好", 4)\
.replace("非常好", 5)\
.replace("无法选择", 0)\
.replace("拒绝回答", 0)
df["D24_float"] = df["D24_float"].astype(float)
df["D24_float"]
```

```python
sns.barplot(x="A30_4_sorted", y="D36_float", data=df)
plt.xticks(rotation=30, ha="right")
plt.ylabel("主观幸福感")
plt.xlabel("阅读行为")
plt.show()
```
